### Restaurants map:

```python
map_rest = Map(location = gf.get_coordenadas("Madrid"), zoom_start = 15)
```

```python
for i,row in restaurants.iterrows():
    dicc = {"location": [row["Latitud"], row["Longitud"]], "tooltip": row["Name"]}
    
    if row["Rating"] >= 4.0:
        icono = Icon(color = "green",
                     prefix="fa",
                     icon="thumbs-o-up",
                     icon_color="black"
        )
    elif row["Rating"] < 4.0 & row["Rating"]>= 2.5:
        icono = Icon(color = "orange",
                     prefix="fa",
                     icon="hand-o-right",
                     icon_color="black")
        
    elif row["Rating"] < 2.5:
        icono = Icon(color = "red",
                     prefix="fa",
                     icon="thumbs-o-down",
                     icon_color="black")
        
    mark = Marker(**dicc, icon=icono)
    mark.add_to(map_rest)
map_rest
```
