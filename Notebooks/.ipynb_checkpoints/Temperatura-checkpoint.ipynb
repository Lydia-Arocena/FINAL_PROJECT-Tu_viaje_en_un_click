#### Limpo las columnas:

```python
def clean_columns(x):
    return x.split("\n")[0]
```

```python
months=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Year"]
for x in months:
    temp[x]=temp[x].apply(clean_columns)
```
